# Add new data points (elevation in m) to database
We decided to use existing data (coordinates) to make new dat points (Elevation in m) so we can use it in our prediction model

```python
# source: https://stackoverflow.com/a/51064850/18261474

# script for returning elevation from lat, long, based on open elevation data
# which in turn is based on SRTM
def get_elevation(lat, long):
    time.sleep(0.05) # added a sleep timer because we would overload the api otherwise
    query = ('https://api.open-elevation.com/api/v1/lookup'
             f'?locations={lat},{long}')
    r = requests.get(query).json()  # json object, various ways you can extract value
    # one approach is to use pandas json functionality:
    elevation = pd.json_normalize(r, 'results')['elevation'].values[0]
    return elevation
```

```python
# test fucntion
get_elevation(0.316231, 32.966182)
```

### chosen features

- `water_source_clean`: Cleaned version of the water source data.
- `management_clean `: Select the classification of the entity that directly manages the water point.
- `age_in_years`: 
- `distance_to_primary`: Distance between water point and nearest primary road per OpenStreetMap
- `distance_to_city`: Distance between water point and nearest city per OpenStreetMap
- `distance_to_town`: Distance between water point and nearest town per OpenStreetMap
- `landcover_code`: Extraction of point values for landcover at each water point GPS location from 2020 dataset.
- `acled_index`: For each water point, the total number of recorded events within a 10 km buffer of its GPS location for the past 10 years (2010 - 2020) is calculated using the Armed Conflict Location and Event Database (ACLED).
- `bgs_dtw`: Extracted point values for each water point GPS location for depth to groundwater (mbgl) using the British Geological Survey Africa Groundwater Atlas.
- `bgs_prod`: Extracted point values for each water point GPS location for groundwater productivity (l/s) using the British Geological Survey Africa Groundwater Atlas.
- `bgs_recharge	`: Extracted point values for each water point GPS location for recharge, the replenishment of groundwater in aquifers by rainfall, using the British Geological Survey Africa Groundwater Atlas.
- `rwi`: This index is an estimate of relative wealth of the people living in each micro-region relative to others in the same country.
- `pressure`: Percent of maximum capacity being utilized based on assigned_population estimate (pressure = assigned_population/usage_cap)
- `criticality`: Ratio of assigned_population to local_population. Indicator of how many alternative functional points are available to poplation.
- `years_since_report`: The number of years since the report date. Calculated by subtracting the report date from today's date.
- `elevation_in_m`:

# Prediction model
## random forest classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Init train data
# X_train = pd.get_dummies(train_df, columns=['water_source_clean','management_clean'])
X_train = train_df
X_train = X_train.drop(columns=['status_id'])

y_train = train_df['status_id']

# Init test data
# X_test = pd.get_dummies(test_df, columns=['water_source_clean','management_clean'])
X_test = test_df
X_test = X_test.drop(columns=['status_id'])
y_test = test_df['status_id']
```

```python
rf_model = RandomForestClassifier(n_estimators=50, random_state=44)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
